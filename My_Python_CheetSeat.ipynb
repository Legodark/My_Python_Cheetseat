### Longitud maxima de una linea de código
---

Se debe de limitar el numero de caracteres por linea a 79

Para los docstrings y comentarios la longitud debe de limitarse a 72 caracteres

Estas limitaciones se pueden configurar dentro de los ajustes de los editores


### Hacer un salto de linea antes o despues del operador
---

Hacer un salto de linea nos ayudara ver mejor las operaciones del código, además de que queda visualmente mas atractivo para nuestra vista.

Esto en python es mas importante puesto que permite hacer un salto de linea aunque el operador se quede arriba u abajo.

![indentación en funciones](https://drive.google.com/uc?id=1-ElhP6CjbmjtKxBydCXw3Feq_Y2XOEXo)


### Espacios en blanco entre lineas
---

Para las funciones y clases debemos de saltar 2 lineas, y para los metodos dentro de una clase debemos saltar 1 linea, como se muestra en el ejemplo:


```python

def top_level_function1():
    pass


def top_level_function2():
    pass


class TestClass(object):
    
    def class_method1():
        pass

    def class_method2():
        pass


class TestClass2(object):
    
    def class2_method1():
        pass

    def class2_method2():
        pass

```
Podemos usar mas saltos de linea en blanco, pero con cuidado.

### Codificación del archivo
---

El nucleo de Python usa UTF-8, por lo que no debería tener una declaración diferente cuando trabajemos con el.

Los identificadores de la libreria estandar de Python deben usar solo identificadores ASCII y deben usar palabras en inglés siempre que sea posible
